```python
dataframes = {}
for file in os.listdir(DATA_PATH):
    name = file.split(".")[0]
    print(name)
    dataframes[name] = pd.read_csv(os.path.join(DATA_PATH, file))
    display(dataframes[name])
```

```python
dataframes["merc"] = dataframes["merc"].append(dataframes["cclass"])
dataframes["ford"] = dataframes["ford"].append(dataframes["focus"])
del dataframes["focus"], dataframes["cclass"]
dataframes["ford"]
```

**Hasta aquÃ­ el dataframe de ford y mercedes no contiene datos duplicados, veamos el resto de dataframes de los otros modelos**

```python
for make in dataframes:

    if make not in ["merc", "ford"]:
        duplicates = dataframes[make].duplicated().sum()

        if duplicates > 0:
            dataframes[make].drop_duplicates(inplace=True)
    if(dataframes[make].duplicated().sum()) != 0:
        display(make) #check
```
